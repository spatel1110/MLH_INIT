```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gas = pd.read_csv('a.csv')

plt.figure(figsize=(8,5))

plt.title('Gas prices in us Dollars')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker = '.')

plt.xticks(gas.Year[::3])
plt.xlabel('Year' )
plt.ylabel('Gas price in us doller')
plt.legend()

plt.show()
```
